```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
df = pd.read_csv('teacher_reviews.csv')
df
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
def analyze_sentiment(text):
    sentiment_scores = sia.polarity_scores(str(text))
    if sentiment_scores['compound'] >= 0.05:
        return 'Positive'
    elif sentiment_scores['compound'] <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
df['Sentiment'] = df['Positive Review'].apply(analyze_sentiment)
df.to_csv('teacher_reviews_with_sentiment.csv', index=False)
```

```python
for index, row in df.iterrows():
    print("Teacher Name:", row['Teacher Name'])
    print("Subject:", row['Subject'])
    print("Positive Review:", row['Positive Review'])
    print("Negative Review:", row['Negative Review'])
    print("Sentiment:", row['Sentiment'])
    print("\n")
```
